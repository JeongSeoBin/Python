# while문
반복처리를 해야 겠는데 몇 번을 반복해야 할지 반복횟수를 정확히 모르는 경우 사용한다

조건이 참인 동안 반복한다. 조건이 거짓이 되면 반복을 종료한다
while 명령의 구조
while 조건식:
   조건식이 참일 경우 실행할 문장
   ...

while 명령의 최초 조건이 거짓일 경우 while문을 한 번도 실행하지 않는다
while 명령의 조건이 항상 참이 될 경우 무한 루트에 빠질 수 있으니 조건을 세울 때 주의해야 하고 의도적으로 무한 루프를 만들어 사용할 경우 반드시 종료 조건을 지정해서 무한반복을 빠져 나갈 수 있는 처리를 해야한다 => 반복의 탈출에는 break 명령을 사용한다

```python
# 변수는 선언된 후 사용되기 전애 반드시 초기화 되어야 한다
i = 0 # 1에서 10까지 1씩 증가하는 변수
total = 0 # 1에서 10까지의 합계를 기억하는 변수
while True:
    i = i + 1
    total = total + i
    if(i == 10):
        break
print('1 ~ 10 까지의 합계 : {}'.format(total))
```

```python
n = int(input("몇까지 더할래?"))
i = 0
total = 0
while i <= n:
    i += 1
    total += i
print('1 ~ {} 까지 합계 : {}'.format(n,total))
```

```python
menu = 0
while menu != 5:
    while True:
        print('*' * 50)
        print("1 입력 2 보기 3 수정 4 삭제 5 종료")
        print('*' * 50)
        menu = int(input('원하는 매뉴를 입력하시오: '))
        if 1 <= menu <= 5:
            break # while 이나 for 명령의 반복 범위를 탈출
        else:
            print('메뉴는 1~5사이의 정수를 입력해야 합니다')
    # 메뉴에 1~5사이의 정수를 입력해서 1 <= menu <=5 조건이 참이면 이곳으로 넘어 온다
    if menu == 1:
        print('입력작업')
    elif menu == 2:
        print('보기작업')
    elif menu == 3:
        print('수정작업')
    elif menu == 4:
        print('삭제작업')
# 메뉴 5가 입력되었으므로 프로그램을 종료한다
print('프로그램이 종료되었습니다')
```

```python
# 유클리드 호제법을 사용하여 최대 공약수와 최소공배수 구하기

# 숫자 2개를 입력받아 큰수, 작은수를 판단한다
a, b = map(int, input('최대 공약수와 최소공배수를 계산할 숫자 2개를 입력하시오 :  ').split())
if a > b:
    high = a; low = b
else:
    high = b; low = a

# 큰 수를 작은 수로 나눈 나머지가 0이 될 때까지 반복한다 => 몇 번을 반복해야 0이 되는지 모른다
# 무한 루프를 반복하다가 큰 수를 작은 수로 나눈 나머지가 0이 되는 순간 무한 루프를 탈출한다
while True:
    # 큰 수를 작은 수로 나눈 나머지를 계산한다
    r = high % low
    # 나머지 0이면 반복을 탈출한다
    if r == 0:
        break
    # 나머지가 0이 아니므로 다음을 계산하기 위해서 큰 수에 작은 수를 넣고 작은 수에 나머지를 넣는디
    high = low
    low = r
    
# 큰 수를 작은 수로 나눈 나머지가 0이 되면 작은 수가 최대공약수가 되고 
# 입력 받은 숫자 2개를 곱한 값을 최대 공약수로 나누면 최소 공뱌수가 된다
print('최대 공약수 : {} 최소 공배수 : {}'.format(low, int(a * b / low)))
```
