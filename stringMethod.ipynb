# len()
인수로 지정된 문자열을 구성하는 문자의 개수

# count()
인수로 지정된 문자열의 출연 횟수

# find() & index()
인수로 지정된 문자열이 처음 나타나는 위치

# join()
인수로 지정된 문자열 사이사이에 '.'앞의 문자열을 삽입

# upper() & lower()
문자열을 무조건 대문자로 변환 & 문자열을 무조건 소문자로 변환

# lstrip() & rstrip() & strip()
문자열의 왼쪽의 불필요한 공백제거 & 오른쪽의 불필요한 공백제거 & 양쪽의 불필요한 공백 및 문자 제거

# replace(old, new)
old 문자열을 new 문자열로 치환

# method 적용방법
len(string)
나머지는 string.fun()
'문자'.join(string)

함수를 적용한다고 원본이 수정되는거 x

```python
string = 'We are the champions, My friend!'

# len()
print(len(string))
print(string[-1])
print(string[len(string) - 1])
print('=' * 75)

# count()
print(string.count('e'))
print(string.count('we'))
print(string.count('We'))
print('=' * 75)

# find()
print(string.find('the'))
print(string.find('The')) # 찾는 문자가 없으면 -1 리턴
print('=' * 75)
# index()
print(string.index('the'))
# print(string.index('The')) # 찾는 문자가 없으면 에러
print('=' * 75)

string = 'TjoeunIT'

# join()
print('/'.join(string))
print('^^;'.join(string))
print('=' * 75)

# upper()
print(string.upper())
# lower()
print(string.lower())
print('=' * 75)

string = '   python   '

# lstrip()
print(string)
print(len(string))
print(string.lstrip())
print(len(string.lstrip()))
# rstrip()
print(string.rstrip())
print(len(string.rstrip()))
# strip()
print(string.strip())
print(len(string.strip()))
print('www.example.com'.strip('w'))
print('www.example.com'.strip('w.com'))
print('=' * 75)

string = 'We are the champions, My friend!'

# replace()
print(string.replace('My','your'))
# replace()를 이용하여 공백제거하기
print('  hello   word  '.strip())
print('  hello   word  '.replace(' ',''))
print('=' * 75)

# split()
# 구분자로 나눈 결과는 list 타입으로 변환된다
print(string.split())
print(type(string.split()))
print('=' * 75)

# join()과 split()를 이용해서 공백문자 2개 이상인 경우에 공백문자를 1개만 나오게 한다
print(' '.join(' hello   world '.split()))
```
