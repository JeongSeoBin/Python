```python
# 미방진
# n행n열의 가로 세로 대각선 어느 방향으로 덧셈을 해도 같은 값이 나온다
# 최초의 1이 채워지는 자리는 첫 줄 정 가운데 열이다 => 5행5열일 경우 0행2열에 1이 채워 진다
# 미방진에 채운 숫자가 n의 배수가 아니면 다음 숫자는 오른쪽 위 방향(행-1 열+1)에 채워진다
# 미방진에 채운 숫자가 n의 배수이면 다음 숫자는 아래줄(행+1)에 채워 진다
# 다음 숫자가 채워질 위치가 리스트의 행 범위를 벗어 나면 마지막 행에, 열 범위를 벗어 나면 0번째 열에 채운다
a = [[0] * 5 for i in range(5)]

# 최초의 1이 채워질 자리
i = 0
j = 2

for k in range(1,26):
    # 미방진에 숫자를 채운다
    a[i][j] = k
    # 채운 숫자가 5의 배수인가 비교해서 다음 숫자가 채워 질 위치를 결정한다
    if k % 5 == 0:
        # 채운 숫자가 5의 배수이면 다음에 숫자가 채워 질 위치는 아랫줄이다
        i += 1
    else:
        # 채운 숫자가 5의 배수가 아니면 다음에 숫자가 채워 지는 위치는 오른쪽 위 방향이다
        i -= 1
        # 행이 리스트의 범위를 벗어나면(-1이 되면)마지막 행으로(4) 변경한다
        if 0 > i :
            i = 4
        j += 1
        # 열이 리스트의 범위를 벗어나면(5가 되면) 첫 번째 열(0)로 변경한다
        if j > 4:
            j = 0
            
for i in range(5):
    for j in range(5):
        print('{0:2d} '.format(a[i][j]), end = '')
    print()
        
```

```python
# 3이상인 홀수를 압력받는다
while True:
    n = int(input('3이상인 홀수를 입력하시오 : '))
    if n >= 3 and n % 2 == 1:
        break
    print('3이상인 홀수를 입력해야 합니다')
    
# n행n열 리스트를 선언한다
a = [[0] * n for i in range(n)]

# 최초의 1이 채워질 위치를 계산한다
i = 0
j = n // 2

for k in range(1, n ** 2 + 1):
    # 미방진에 숫자를 채운다
    a[i][j] = k
    # 채운 숫자가 5의 배수인가 비교해서 다음 숫자가 채워 질 위치를 결정한다
    if k % n == 0:
        # 채운 숫자가 5의 배수이면 다음에 숫자가 채워 질 위치는 아랫줄이다
        i += 1
    else:
        # 채운 숫자가 5의 배수가 아니면 다음에 숫자가 채워 지는 위치는 오른쪽 위 방향이다
        i -= 1
        # 행이 리스트의 범위를 벗어나면(-1이 되면)마지막 행으로(4) 변경한다
        if 0 > i :
            i = n - 1
        j += 1
        # 열이 리스트의 범위를 벗어나면(5가 되면) 첫 번째 열(0)로 변경한다
        if j > n - 1:
            j = 0
            
for i in range(n):
    for j in range(n):
        print('{0:2d} '.format(a[i][j]), end = '')
    print()
```
