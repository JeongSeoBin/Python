# 폴더안에 있는 모듈 import
import할 모듈이 현재 작업중인 프로그램이 위치한 폴더에 있으면 import뒤에 모듈이름만 적어 주면 되지만
import할 모듈이 현재 폴더 아래의 폴더에 위치한 경우에는 폴더이름.모듈이름과 같은 형태로 import해야 한다

```python
import lib.selectionSortModuleLib as ssm

data = [8, 3, 4, 9, 1, 100]
data = ssm.selectionSortAsc(data)
print('오름차순 정렬 결과 : {}'.format(data))
data = ssm.selectionSortDesc(data)
print('내림차순 정렬 결과 : {}'.format(data))
```
